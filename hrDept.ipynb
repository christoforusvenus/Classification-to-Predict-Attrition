## Divide Feature into Categorical and Numerical ##

```python
df.hist(bins=30, figsize=(20, 20), color='b')
```

```python
catCols = [col for col in df.select_dtypes(include='object').columns.tolist()] + ['Education', 
                                                                                           'EnvironmentSatisfaction', 
                                                                                           'JobInvolvement', 'JobLevel', 
                                                                                           'JobSatisfaction', 
                                                                                           'RelationshipSatisfaction', 
                                                                                           'StockOptionLevel', 
                                                                                           'WorkLifeBalance', 
                                                                                           'PerformanceRating']
```

```python
numCols = [col for col in df.columns if col not in catCols]
```

### Categorical Feature Handling ###

```python
catDf = df[catCols].copy()
```

```python
catDf.head()
```

```python
for col in catDf.columns:
    print(f"Value counts for {col}:\n{catDf[col].value_counts()}\n")
```

#### Drop Feature that only has 1 value and feature PerformanceRating which i assume this as an ordinal but it has missing value which not start form 1 and only provided number 3 & 4. #### 
#### I assume this could make my model confuse when there are user that has PerformanceRating except 3 / 4 so i will taken this out ####

```python
catDf.drop(["Over18", "PerformanceRating"], axis=1, inplace=True)
```

```python
catDf.head()
```

```python
catDf['Attrition'] = catDf['Attrition'].apply(lambda x:1 if x == "Yes" else 0)
#This is the target which will be taken out later
```

```python
catDf['OverTime'] = catDf['OverTime'].apply(lambda x:1 if x == "Yes" else 0)
#Change the value into 0 and 1 beacuse it has boolean value
```

```python
catDf['BusinessTravel'] = catDf['BusinessTravel'].apply(lambda x: 2 if x == 'Travel_Frequently' else (1 if x == 'Travel_Rarely' else 0))
#Label encoding because it is ordinal 
```

```python
catDf.head()
```

```python
catDf = pd.get_dummies(catDf, columns=['Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus'], 
                       prefix=['Dept', 'EduField', 'Gender', 'JobRole', 'MaritalStatus'])
#One Hot Encoding for Department, EducationField, Gender, JobRole, and MaritalStatus because it's nominal
```

```python
catDf.head()
```

### Numerical Feature Handling ###

```python
numDf = df[numCols].copy()
```

```python
numDf.head()
```

```python
for col in numDf.columns:
    unique_values = numDf[col].nunique()
    if unique_values <= 5:
        print(f"Value counts for {col} (Unique Values: {unique_values}):\n{numDf[col].value_counts()}\n")
#show feature that only has 1 unique value, i use 5 as threshold 
```

```python
numDf.drop(['EmployeeCount', 'StandardHours'],axis=1, inplace=True)
#Drop feature that only has 1 unique value
```

```python
numDf['EmployeeNumber'].nunique()
```

```python
numDf.drop('EmployeeNumber', axis=1, inplace=True)
#this feature has no information for the model because it's like the ID of the user which all are unique
```

#### Show Correlation and Drop feature that has high correlation > 0.7 ####

```python
correlation_matrix = numDf.corr()

plt.figure(figsize=(20, 20))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
corr_matrix = numDf.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop_hicorr = [column for column in upper.columns if any(upper[column] > 0.7)]
```

```python
to_drop_hicorr
```

```python
numDf.drop(to_drop_hicorr, axis=1, inplace=True)
```

```python
numDf.head()
```
